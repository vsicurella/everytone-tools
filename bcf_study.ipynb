```python
import numpy as np
from utils import *
import matplotlib.pyplot as plt
```

```python

MAX_TWOS=64

num = 2*2*3*5*5
for i in range(num):
    dec = 1.0 + float(i)/num
    bcf = get_bcf_until(dec, MAX_TWOS)

    term_k = len(bcf)
    if term_k >= MAX_TWOS:
        for k in range(term_k):
            if k > 64:
                break
            if bcf[term_k - k - 1] != 2:
                term_k -= k - 1
                break
    
    bcf_term = bcf[:term_k]
    bcf_string = ", ".join([str(d) for d in bcf_term])
    if term_k < len(bcf) or len(bcf) == MAX_TWOS:
        bcf_string += ",..."

    print(f'{dec:.6f}:\t{bcf_string}')
        
```
